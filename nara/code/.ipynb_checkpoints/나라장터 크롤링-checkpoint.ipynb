```python
# 크롬 브라우저를 띄우기 위해, 웹드라이버를 가져오기
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

from selenium.webdriver.common.by import By

import os
import sys

if getattr(sys, 'frozen', False):
    chromedriver_path = os.path.join(sys._MEIPASS, "chromedriver.exe")
    driver = webdriver.Chrome(chromedriver_path)
else:
    driver = webdriver.Chrome(r'C:\chromedriver.exe')

# driver = webdriver.Chrome(r'C:\chromedriver.exe')
driver.get('https://www.g2b.go.kr/pt/menu/selectSubFrame.do?framesrc=/pt/menu/frameTgong.do?url=https://www.g2b.go.kr:8101/ep/tbid/tbidFwd.do?bidSearchType=1')
time.sleep(1)

driver.switch_to.frame('sub')
driver.switch_to.frame('main')
check_box_1 = driver.find_element(By.XPATH, '//*[@id="exceptEnd"]')
check_box_1.click()
time.sleep(2)

search_button_1 = driver.find_element(By.XPATH, '//*[@id="search"]/table/tbody/tr[7]/td/div/button[1]')
search_button_1.click()
time.sleep(2)

driver.switch_to.window(driver.window_handles[1])
driver.find_element(By.XPATH, '//*[@id="industrialCd"]').click()
time.sleep(1)

elements = driver.find_element(By.XPATH, '//*[@id="industrialCd"]')
elements.send_keys('4997')
time.sleep(1)

driver.find_element(By.XPATH , '//*[@id="bt_search"]/span').click()
time.sleep(1)

driver.find_element(By.XPATH, '//*[@id="ebid"]/div[5]/table/tbody/tr/td[4]/a').click()
time.sleep(1)

driver.switch_to.window(driver.window_handles[0])
driver.switch_to.frame('sub')
driver.switch_to.frame('main')
search_button_2 = driver.find_element(By.XPATH, '//*[@id="buttonwrap"]/div/a[1]')
search_button_2.click()
time.sleep(2)

# table = driver.find_element_by_class_name('results')
# thead = table.find_element_by_tag_name("thead")
# rows = thead.find_elements_by_tag_name("tr")

table = driver.find_element(By.CLASS_NAME, 'results')
thead = table.find_element(By.TAG_NAME, 'thead')
rows = thead.find_elements(By.TAG_NAME, 'tr')

col_name = []

for index, value in enumerate(rows) :
    for i in range(len(value.find_elements(By.TAG_NAME, "th"))):
        col_name.append(value.find_elements(By.TAG_NAME, "th")[i].text)

df = pd.DataFrame(columns = col_name)

# tbody = table.find_element_by_tag_name("tbody")
# rows = tbody.find_elements_by_tag_name("tr")

tbody = table.find_element(By.TAG_NAME, 'tbody')
rows = tbody.find_elements(By.TAG_NAME, 'tr')

row_data = []

for index, value in enumerate(rows) :
    for i in range(len(value.find_elements(By.TAG_NAME, "div"))) :
        row_data.append(value.find_elements(By.TAG_NAME, "div")[i].text)

for i in range(len(row_data) // 10):
    df.loc[i] = row_data[i*10:(i+1)*10]

driver.quit()

import datetime
d_today = datetime.date.today()
path = '../data/'
df.to_csv(path+str(d_today)+'.csv', index = False, encoding='cp949')
```
